```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.linspace(0, 1, 100)
```

```python
y_phi_5_over_10 = (1/(1-x+x*0.5))**2
```

```python
plt.style.use('ggplot')
```

```python
plt.plot(x,y_phi_5_over_10)
plt.xlabel("$f$")
plt.ylabel("$K{p^*}^2$")
plt.grid(False)
plt.tick_params(axis='both', 
                which='both', 
                bottom='off', 
                top='off', 
                labelbottom='off', 
                right='off', 
                left='off', 
                labelleft='off')
plt.title("Incentive to adjust price as function of $f$, with $\phi < 1$")
plt.tight_layout()
plt.savefig("phi_less_than_one.pdf")
plt.show()
```

```python
y_phi_15_over_10 = (1/(1-x+x*1.5))**2
plt.plot(x,y_phi_15_over_10)
plt.xlabel("$f$")
plt.ylabel("$K{p^*}^2$")
plt.grid(False)
plt.tick_params(axis='both', 
                which='both', 
                bottom='off', 
                top='off', 
                labelbottom='off', 
                right='off', 
                left='off', 
                labelleft='off')
plt.title("Incentive to adjust price as function of $f$, with $\phi > 1$")
plt.tight_layout()
plt.savefig("phi_greater_than_one.pdf")
plt.show()
```
